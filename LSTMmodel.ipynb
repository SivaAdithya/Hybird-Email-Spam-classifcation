```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow import keras
from keras.preprocessing.text import Tokenizer
from keras import preprocessing
from keras.models import Sequential
from keras.layers import Flatten, Dense
from keras.layers import Embedding,LSTM,Dropout
from keras.utils import pad_sequences
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('D:\Study material\sem6\AI for CyberSecurity\Jcomp\Data\spam_or_not_spam.csv')
data.head()
```

```python
data.isnull().sum()
```

```python
data.dropna(inplace = True)
```

```python
data.isnull().sum()
```

```python
data.head()
```

```python
text = []
for i in data['email']:
    text.append(i)
text[0:5]
```

```python
len(text)
```

```python
tokenizer = Tokenizer(num_words=1500)
tokenizer.fit_on_texts(text)
sequences = tokenizer.texts_to_sequences(text)

x = sequences
y = data['label']
```

```python
x[0:1]
```

```python
y[:5]
```

```python
y.unique()
```

```python
mx = 0
for i in range(len(x)):
    if(len(x[i])>mx):
        mx = i
mx
```

```python
x = pad_sequences(x, maxlen=mx)
len(x[0])
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size = 0.8)
```

```python
model = Sequential()
model.add(Embedding(1500,128,input_length=2744))
lstM = 256
model.add(LSTM(lstM))
model.add(Dense(32, activation = 'relu'))
model.add(Dropout(0.2))
model.add(Dense(8, activation = 'relu'))
model.add(Dropout(0.1))
model.add(Dense(1, activation = 'sigmoid'))
model.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
model.summary()
```

```python
history = model.fit(x_train,y_train,epochs=5, validation_split = 0.15)
```

```python
model.save('lstm.h5')
```

```python
model.evaluate(x_test,y_test)
```
